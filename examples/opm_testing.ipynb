```python
## Example Query Strings

no_tilt_sql = """
    WHERE tilt=0
    """
has_tilt_sql = """
    WHERE tilt>0
    """      
channel_sql =     """
    WHERE (name LIKE '%channel%' OR name LIKE '%porin%') 
    AND NOT (name LIKE '%domain%' OR name LIKE '%monomer%') 
    AND (Class GLOB 'Beta-barrel transmembrane' )
    """

## Construct queried dataframe

df_channels = opm.sql_search(df_proteins,"*",channel_sql)

df_no_tilt = opm.sql_search(df_proteins,"*",no_tilt_sql)

df_has_tilt = opm.sql_search(df_proteins,"*",has_tilt_sql)


## save queries

query_dict = opm.add_query(df_channels,"channels",query_dict)
query_dict = opm.add_query(df_no_tilt,"no tilt",query_dict)
query_dict = opm.add_query(df_has_tilt,"has tilt",query_dict)


## print saved queries

query_dict.keys()
```
