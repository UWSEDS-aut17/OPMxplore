```python
import pandas as pd
import numpy as np
import pypdb as pdb
#be sure to pip install xlrd (an optional package for pandas to read excel files)
```

```python
#Read the transmembrane protein data (in excel format) extracted from the website: http://opm.phar.umich.edu/types.php
transmembrane = pd.read_excel("1_Transmembrane_type.xlsx", "Sheet1")
transmembrane
```

```python
##Read the monotopic protein data (in excel format) extracted from the website: http://opm.phar.umich.edu/types.php
monotopic = pd.read_excel("2_Monotopic_type.xlsx", "Sheet1")
monotopic
```

```python
#Read the peptides data (in excel format) extracted from the website: http://opm.phar.umich.edu/types.php
peptides = pd.read_excel("3_Peptides_type.xlsx", "Sheet1")
peptides
```

```python
# To have continuous index numbers in the merged df 
allprotein_df = pd.concat([transmembrane, monotopic, peptides], keys =["transmembrane", "monotopic", "peptides"], ignore_index=True)
#allprotein_df.to_csv('allprotein.csv', index=False)
```

```python
#Concatenate the 3 dataframes together to get a master list of proteins, while keeping their identity keys 
allprotein_df = pd.concat([transmembrane, monotopic, peptides], keys =["transmembrane", "monotopic", "peptides"])
```

```python
# Find the average hydrophobic thickness or Depth of the different membranes
allprotein_df.groupby('Localization')['Hydrophobic_Thickness or_Depth '].mean()
```

```python
# other averages group by the localization
allprotein_df.groupby('Localization').mean()
```

```python
# Find the average hydrophobic thickness and number of subunits the different membranes to plot
allprotein_df.groupby('Localization')['Hydrophobic_Thickness or_Depth ', 'Num_Subunits'].mean()
```
